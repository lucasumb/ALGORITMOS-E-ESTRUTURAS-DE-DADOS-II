##Exercicio 2.1

Nesta primeira parte, ele pede para calcularmos as probabilidades referente ao cara e coroa. Por exemplo:

p1(H) -> probabilidade da primeira moeda ser cara

p2(H) -> probabilidade da segunda moeda ser cara

H-H -> probabilidade da primeira e segunda moeda serem cara

H-T	-> probabilidade da primeira ser cara e segunda moeda ser coroa

e assim por diante...


```python
lista = [(0.5,0.5), (0.6,0.7), (0.4,0.8), (0.1,0.2), (0.3,0.4)]

for value in lista:
  print("p1(H)    p2(H)   H-H   H-T   T-H   T-T")
  hh = value[0]*value[1]
  ht = value[0]*(1-value[1])
  th = (1-value[0])*value[1]
  tt = (1-value[0])*(1-value[1])
  print(f'{value[0]}    {value[1]}    {hh:.2f}    {ht:.2f}    {th:.2f}  {hh:.2f}')
```

##Exercicio 2.2

60% dos emails que chegam são spam, logo 40% não são spam. Há tambem um filtro de spam de 98% de precisão.

P(spam) = 0.6

P(nao spam) = 0.4

P(filtro diz que é spam | é spam) = 0.98

P(filtro diz que é spam | não é spam) = 1 - P(filtro diz que é spam | é spam) = 0.02

Agora, queremos saber a probabilidade de que um e-mail que foi marcado como spam não seja spam, ou seja, P(nao é spam ∣ filtro diz que é spam).
Usando o Teorema de Bayes:

P(nao é spam ∣ filtro diz que é spam) = (P(filtro diz que é spam | não é spam) * P(nao spam) / P(spam)

No fim, usamos o Teorema de Bayes para calcular quantos desses 963 e-mails deveria suspeitar que não são spam.

963 * P(nao é spam ∣ filtro diz que é spam) = 12.84

```python
P_spam = 0.6

P_nao_spam = 0.4

P_filtro_diz_que_é_spam_e_é_spam = 0.98

P_filtro_diz_que_é_spam_e_nao_é_spam = 1 - P_filtro_diz_que_é_spam_e_é_spam

P_nao_é_spam_filtro_diz_que_é_spam = P_filtro_diz_que_é_spam_e_nao_é_spam * P_nao_spam / P_spam

resposta = 963 * P_nao_é_spam_filtro_diz_que_é_spam

print(f'A probabilidade de que um e-mail que foi marcado como spam não seja spam é {resposta:.2f}')
```

##Exercicio 2.4

Nesta questão ele pede para desenhar a função de massa de probabilidade e a distribuição cumulativa

```python
import matplotlib.pyplot as plt
import numpy as np

outcomes = np.array([1, 2, 3, 4, 5, 6, 7, 8])
probabilities = np.array([0.1, 0.15, 0.2, 0.21, 0.17, 0.09, 0.06, 0.02])

cumulative_probabilities = np.cumsum(probabilities)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

ax1.bar(outcomes, probabilities, color='skyblue', edgecolor='black')
ax1.set_title('Probability Mass Function (PMF)')
ax1.set_xlabel('Outcome')
ax1.set_ylabel('Probability')

ax2.bar(outcomes, cumulative_probabilities, color="#e41a1c", edgecolor='black')
ax2.set_title('Cumulative Distribution Function (CDF)')
ax2.set_xlabel('Outcome')
ax2.set_ylabel('Cumulative Probability')

plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Criando o grafo
graph = nx.Graph()

# Adicionando nós e arestas
nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
edges = [
    ('A', 'B'),
    ('A', 'G'),
    ('B', 'C'),
    ('B', 'D'),
    ('C', 'D'),
    ('D', 'E'),
    ('D', 'F'),
    ('D', 'G'),
    ('D', 'H'),
]
graph.add_nodes_from(nodes)
graph.add_edges_from(edges)

# Desenhando a rede de correlação
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(graph)
nx.draw(graph, pos, with_labels=True, node_color='skyblue', node_size=1500, font_size=12, font_weight='bold', edge_color='gray')
plt.title("Rede de Correlação")
plt.show()
```

```python
import networkx as nx

def hypergraph_to_unipartite(hypergraph_file):
  """Transforms a hypergraph into a unipartite graph."""
  G = nx.Graph()
  with open(hypergraph_file, 'r') as f:
    for line in f:
      nodes = line.strip().split()
      for i in range(len(nodes)):
        for j in range(i + 1, len(nodes)):
          G.add_edge(nodes[i], nodes[j])
  return G

def hypergraph_to_bipartite(hypergraph_file):
  """Transforms a hypergraph into a bipartite graph."""
  G = nx.Graph()
  with open(hypergraph_file, 'r') as f:
    hyperedge_id = 0
    for line in f:
      nodes = line.strip().split()
      hyperedge_node = f"hyperedge_{hyperedge_id}"
      G.add_node(hyperedge_node, bipartite=0)
      for node in nodes:
        G.add_node(node, bipartite=1)
        G.add_edge(hyperedge_node, node)
      hyperedge_id += 1
  return G

# Transform data3.txt into a unipartite graph
unipartite_graph = hypergraph_to_unipartite("data3.txt")
print("Unipartite Graph:")
print(unipartite_graph.edges())

# Transform data3.txt into a bipartite graph
bipartite_graph = hypergraph_to_bipartite("data3.txt")
print("\nBipartite Graph:")
print(bipartite_graph.edges())

```

```python
import networkx as nx

def aggregate_dynamic_network(edgelist_file, window_size):
  """Aggregates a dynamic network using a disjoint window."""

  G_aggregated = nx.Graph()
  edges_in_window = {}

  with open(edgelist_file, 'r') as f:
    for line in f:
      node1, node2, start_snapshot, end_snapshot = line.strip().split()
      start_snapshot = int(start_snapshot)
      end_snapshot = int(end_snapshot)

      for snapshot in range(start_snapshot, end_snapshot + 1):
        window_start = (snapshot // window_size) * window_size
        window_end = window_start + window_size - 1

        if window_start <= snapshot <= window_end:
          edge_key = (node1, node2, window_start)
          if edge_key not in edges_in_window:
            edges_in_window[edge_key] = 0
          edges_in_window[edge_key] += 1

  for (node1, node2, window_start), edge_count in edges_in_window.items():
      if edge_count > 0:
          G_aggregated.add_edge(node1, node2)

  return G_aggregated



# Aggregate the network using a disjoint window of size 3
aggregated_graph = aggregate_dynamic_network("data4.txt", 3)

print("Aggregated Graph:")
print(aggregated_graph.edges())


```
